#### __Q.2) What is the average age of men?__

```python
average=round(final_df.loc[final_df['sex']=='Male', 'age'].mean(),2)
print("The average age of men is",average)
```

#### __Q.3) What is the percentage of people who have a Bachelor's degree?__

```python
total=final_df ['education'].count()
bachelors=final_df.loc[final_df['education']=='Bachelors', 'education'].count()
people_w_bachelors=round((bachelors/total)*100,2)
print(people_w_bachelors,"% of people have a Bachelor's degree")
```

#### __Q.4) What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?__

```python
higher_education=len(df.loc[(df['education'].isin(['Bachelors', 'Masters', 'Doctorate']))])
higher_education_w_salary = df.loc[(df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])) & (df['salary'].str.startswith('>')), 'education'].count()
higher_education_rich = round(higher_education_w_salary*100/higher_education, 1)
print(higher_education_rich,"% of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K")
```

#### __Q.5) What percentage of people without advanced education make more than 50K?__

```python
lower_education=len(df.loc[(~df['education'].isin(['Bachelors', 'Masters', 'Doctorate']))])
lower_education_w_salary = df.loc[(~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])) & (df['salary'].str.startswith('>')), 'education'].count()
lower_education_rich = round(lower_education_w_salary*100/lower_education, 1)
print (lower_education_rich,"% of people without advanced education make more than 50K")
```
