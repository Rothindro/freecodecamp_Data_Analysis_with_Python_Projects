#### __Q.1) Add an overweight column to the data. To determine if a person is overweight, first calculate their BMI by dividing their weight in kilograms by the square of their height in meters. If that value is > 25 then the person is overweight. Use the value 0 for NOT overweight and the value 1 for overweight.__

```python
cfile['overweight']=(cfile['weight']/((cfile['height']/100)**2)).apply(lambda x: 1 if x > 25 else 0)
cfile.head(10)
```

#### __Q.2) Normalize the data by making 0 always good and 1 always bad. If the value of cholesterol or gluc is 1, make the value 0. If the value is more than 1, make the value 1.__

```python
cfile['cholesterol']=cfile['cholesterol'].apply(lambda x: 0 if x < 2 else 1)
cfile['gluc']=cfile['gluc'].apply(lambda x: 0 if x < 2 else 1)
cfile.head(10)
```

#### __Q.3) Convert the data into long format and create a chart that shows the value counts of the categorical features using seaborn's catplot(). The dataset should be split by Cardio so there is one chart for each cardio value. The chart should look like examples/Figure_1.png.__

```python
cfile_gr=pd.melt(cfile, id_vars=["cardio"], value_vars=["active","alco", "cholesterol", "gluc", "overweight", "smoke"])
cfile_gr['total']=1
cfile_gr=cfile_gr.groupby (['cardio','variable','value'], as_index=False).count()
print(cfile_gr)
fig=sns.catplot(x='variable', y='total', data=cfile_gr, hue='value', kind='bar', col='cardio').fig
#fig.savefig('catplot.png')
```

#### __Q.4) Clean the data. Filter out the following patient segments that represent incorrect data:__

-diastolic pressure is higher than systolic (Keep the correct data with (df['ap_lo'] <= df['ap_hi'])) -height is less than the 2.5th percentile (Keep the correct data with (df['height'] >= df['height'].quantile(0.025))) -height is more than the 97.5th percentile -weight is less than the 2.5th percentile -weight is more than the 97.5th percentile

```python
cfile_heat=cfile[(cfile['ap_lo']<=cfile['ap_hi']) & (cfile['height'] >= cfile['height'].quantile(0.025)) & 
(cfile['height'] <= cfile['height'].quantile(0.975)) & (cfile['weight'] >= cfile['weight'].quantile(0.025)) & 
(cfile['weight'] <= cfile['weight'].quantile(0.975))]
cfile_heat
```

#### __Q.5) Create a correlation matrix using the dataset. Plot the correlation matrix using seaborn's heatmap(). Mask the upper triangle. The chart should look like examples/Figure_2.png.__

```python
corr=cfile_heat.corr(method='pearson')
corr
```

```python
import numpy as np
mask=np.triu(corr)
mask
```

```python
import matplotlib.pyplot as plt
fig, ax=plt.subplots(figsize=(14,14))
sns.heatmap(corr, linewidths=1, annot=True, square=True, mask=mask, fmt='.2f', center=0.2, cbar_kws={'shrink':0.7})
```
